### Subteam 1: Creating a dictionary out of the review text file.

Here is Subteam 1's results. In our portion of the project, we are making a dictionary out of the provided text file and in combination with LIWC's dictionary, we will use this to create NLP models later on.

```python

#Meta data file 
#this is the array that will hold the meta data from the file
commentsMeta=[]
ynlist = []

#opening the meta data file
with open("/content/gdrive/MyDrive/CSE-CSB_Capstone/Subteam-1/output_meta_yelpHotelData_NRYRcleaned.txt","r",newline="\n") as fil:
  for comment in fil:
    commentsMeta.append((comment.rstrip()).split(" "))

#replace each of the comments 'Y' or 'N' values with 1 or 0
for comment in commentsMeta:
  if(comment[4] == 'N'):
    comment[4] = '0'
  elif(comment[4] == 'Y'):
    comment[4] = '1'
  ynlist.append(comment[4])

#create a numpy array from the lists
#this will be stacked with out feature extraction array
ynarray = np.asarray(ynlist)

yesindexes = []

for i in range(0,len(ynlist)):
    if int(ynlist[i]) == 1:
      yesindexes.append(i)

#testing the function with print statements
print("First line of the commentsMeta file stored as an array:",commentsMeta[0])
print("Length of the yes and no numpy array: ", len(ynarray))
#print("All indexes where the review was classified as yes: ", yesindexes)
print("Number of indexes where the review was a yes:", len(yesindexes))
```
